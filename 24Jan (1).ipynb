```python
import pandas as pd
import numpy as np
data=pd.read_csv("clevelanda (1).csv")
data.info()
```

```python
data.shape
```

```python
data
```

```python
for column in data.columns:
    print("Unique values for column",column)
    print(data[column].unique())
```

```python
data.replace('?', np.NaN)
```

```python
data["ca"]=pd.to_numeric(data["ca"], errors='coerce')
print(data["ca"])
data["thal"]=pd.to_numeric(data["thal"], errors='coerce')
print(data["thal"])
```

```python
print(data["ca"].mean())
```

```python
ca_mean=data["ca"].mean()
print(ca_mean)
ca_median=(data["ca"].median())
print(ca_median)
ca_mode=data["ca"].mode()[0]
print(ca_mode)
thal_mean=data["thal"].mean()
print(thal_mean)
thal_median=(data["thal"].median())
print(thal_median)
thal_mode=(data["thal"].mode())[0]
print(thal_mode)
```

```python
data['ca']=data['ca'].fillna(ca_mode)
data['thal']=data['thal'].fillna(thal_mode)
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
#univariate analysis

#histogram:shows data distribution
#box plot:highlights summary statistics(mean,median,mode)
#bar chart:displays frequency of categories

#bivariate analysis

#scatter plot:visualizes relation between two continuous variables
#line plot:tracks changes in two related variables over time
#heatmap:shows correlation in matrix format
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = x*2

# first plot with X and Y data
plt.plot(x, y, marker='o')

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('Simple plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Jan', 'Feb', 'March', 'April', 'May', 'June'])
y = np.array([10000,10500,9000,8500,12000,11000])

# first plot with X and Y data
plt.plot(x, y, marker='o', linestyle='dashed', color='green')

plt.xlabel("Months")
plt.ylabel("Savings amt")
plt.title('Savings from January to June')
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['violet','indigo','blue','green','yellow'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Categories vs Values')
plt.show()
```

```python
categories=[22,25,30,35,40,45,50]
values=[10,20,30,40,50]
plt.hist(categories,bins=values,color='violet',edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Categories vs Values')
plt.show()
```

```python
lang=['python','java','c','javascript','c++']
per=[23,30,20,17,10]
plt.pie(per,labels=lang,autopct='%1.1f%%')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Categories vs Values')
plt.legend()
plt.show()
```
