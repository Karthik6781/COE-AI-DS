```python
import pandas as pd
jobs=pd.read_csv("data job posts.csv")
jobs.head()
```

```python
#Check for NaN (missing values) in all columns.
jobs.isna().sum()
```

```python
#Drop irrelevant columns (e.g., Attach, Notes if they contain too much noise).
jobs=jobs.drop(columns=['Notes','Attach'])
jobs.head()
```

```python
print(jobs['date'].mode()[0])
print(jobs['StartDate'].mode()[0])
print(jobs['OpeningDate'].mode()[0])
print(jobs['Deadline'].mode()[0])
```

```python
#Convert date, StartDate, OpeningDate, Deadline into datetime format.
jobs['date']=pd.to_datetime(jobs['date'],errors='coerce',format='%Y-%m-%d')
jobs['StartDate']=pd.to_datetime(jobs['StartDate'],errors='coerce')
jobs['OpeningDate']=pd.to_datetime(jobs['OpeningDate'],errors='coerce')
jobs['Deadline']=pd.to_datetime(jobs['Deadline'],errors='coerce')
jobs.dtypes
```

```python
print(jobs['date'].mode()[0])
print(jobs['StartDate'].mode()[0])
print(jobs['OpeningDate'].mode()[0])
print(jobs['Deadline'].mode()[0])
```

```python
#Fill missing values using mode (for categorical) and median (for numeric) where appropriate.
cat_jobs= jobs.select_dtypes(include=['object']).columns
for i in cat_jobs:
        mode=jobs[i].mode()[0]
        jobs[i].fillna(mode,inplace=True)
jobs.isna().sum()
```

```python
cat_jobs= jobs.select_dtypes(include=['datetime']).columns
for i in cat_jobs:
        mode=jobs[i].mode()[0]
        jobs[i].fillna(mode,inplace=True)
jobs.isna().sum()
```

```python
#Days left to apply â†’ Deadline - OpeningDate
jobs['DaysLeftToApply']=jobs['Deadline'] - jobs['OpeningDate']
#Job start delay â†’ StartDate - OpeningDate.
jobs['JobStartDelay']=jobs['StartDate'] - jobs['OpeningDate']
jobs[['date','StartDate','OpeningDate','Deadline','DaysLeftToApply','JobStartDelay']]
```
